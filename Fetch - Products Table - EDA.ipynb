```python
#Data Loading
import pandas as pd
products_data = pd.read_csv("C:/Users/LENOVO/Downloads/PRODUCTS_TAKEHOME.csv")
```

# Calculating the percentage of missing values in each column


```python
# Calculating the percentage of missing values in each column

missing_values_count = products_data.isnull().sum()

missing_values_percentage = (products_data.isnull().sum() / len(products_data)) * 100

print("Missing Values Count:\n", missing_values_count)
print("\nMissing Values Percentage:\n", missing_values_percentage)
```

# Calculating the percentage of placeholder value in the dataset


```python
# Calculating the percentage of placeholder value in the dataset

placeholder_count = products_data['MANUFACTURER'].value_counts().get('PLACEHOLDER MANUFACTURER', 0)

total_count = len(products_data)
placeholder_percentage = (placeholder_count / total_count) * 100

print(f"Count of 'PLACEHOLDER MANUFACTURER': {placeholder_count}")
print(f"Percentage of 'PLACEHOLDER MANUFACTURER': {placeholder_percentage:.2f}%")
```

## Checking the top 5 most frequent values


```python
## Checking the top 5 most frequent values
import matplotlib.pyplot as plt
products_columns = ['CATEGORY_1', 'CATEGORY_2', 'CATEGORY_3', 'CATEGORY_4', 'MANUFACTURER', 'BRAND']


for column in products_columns:
    top_values = products_data[column].value_counts().nlargest(5)
    
    # Plotting
    plt.figure(figsize=(10, 5))
    top_values.plot(kind='bar', color='skyblue')
    plt.title(f'Top 5 Most Occurring Values in {column}')
    plt.ylabel('Frequency')
    plt.xlabel('Values')
    plt.xticks(rotation=45) 
    plt.show()
```

# Calculating the percentage of 'brand not known'


```python
# Calculating the percentage of 'brand not known'
brand_not_known_count = products_data['BRAND'].value_counts().get('BRAND NOT KNOWN', 0)
total_brand_count = len(products_data['BRAND'])
brand_not_known_percentage = (brand_not_known_count / total_brand_count) * 100
print(f"Percentage of 'brand not known' in the BRAND column: {brand_not_known_percentage:.2f}%")
```
